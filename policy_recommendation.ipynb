```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
policy_data=pd.read_csv("policy.csv")
```

```python
policy_data
```

```python
rating_data=pd.read_csv("ratings.csv")
```

```python
rating_data=rating_data[['user_id', 'policy_id', 'rating', 'timestamp']]
rating_data
```

```python
#grouping by the user id and taking the average rating provided by that user 
Mean = rating_data.groupby(by="user_id",as_index=False)['rating'].mean()
Mean
```

```python
#merging the mean rating with the original dataframe
rating_avg = pd.merge(rating_data,Mean,on='user_id')
rating_avg
```

```python
# finding out the avg rating of each row
rating_avg['avg_rating']=rating_avg['rating_x']-rating_avg['rating_y']
rating_avg.head()
```

```python
#converting the average rating column in 2D table 
```

```python
final=pd.pivot_table(rating_avg,values='avg_rating',index='user_id',columns='policy_id')
```

```python
final
```

```python
# similary creating the pivot table for rating which is actualy assigned by the user
check_table = pd.pivot_table(rating_avg,values='rating_x',index='user_id',columns='policy_id')
check_table.head()
```

```python
# creating first variable using fillna functionn(column wise) for filling the mean value in the final dataframe containing pivot table of average rating
#its like for each policys
```

```python
final_policy = final.fillna(final.mean(axis=0))
```

```python
final_policy
```

```python
# creating second variable using fillna functionn(row wise) for filling the mean value in the final dataframe containing pivot table of average rating
#for each user
final_user = final.apply(lambda row: row.fillna(row.mean()), axis=1)
```

```python
final_user
```

```python
#for each user performing the cosine similarity with each other and  storing in the variable namer similarity_with_user 
b = cosine_similarity(final_user)
np.fill_diagonal(b, 0 )
similarity_with_user = pd.DataFrame(b,index=final_user.index)
similarity_with_user.columns=final_user.index
similarity_with_user.head()
```

```python
#for each policy performing the cosine similarity with other policies and  storing in the variable namer similarity_with_policy
cosine = cosine_similarity(final_policy)
np.fill_diagonal(cosine, 0 )
similarity_with_policy =pd.DataFrame(cosine,index=final_policy.index)
similarity_with_policy.columns=final_user.index
similarity_with_policy.head()
```

```python
def find_n_neighbours(df,n):
    order = np.argsort(df.values, axis=1)[:, :n]
    df = df.apply(lambda x: pd.Series(x.sort_values(ascending=False)
           .iloc[:n].index, 
          index=['top{}'.format(i) for i in range(1, n+1)]), axis=1)
    return df
```

```python
#now going to use knearest neighbours to find out
# top 40 neighbours for each user
sim_user_40_u = find_n_neighbours(similarity_with_user,40)
sim_user_40_u
```

```python
#now going to use knearest neighbours to find out
# top 30 neighbours for each policy
sim_user_40_m = find_n_neighbours(similarity_with_policy,40)
sim_user_40_m
```

```python

def get_user_similar_policys( user1, user2 ):
    common_policys = rating_avg[rating_avg.user_id == user1].merge(
    rating_avg[rating_avg.user_id == user2],
    on = "policy_id",
    how = "inner" )
    return common_policys.merge( policy_data, on = 'policy_id' )
```

```python
def User_item_score(user,item):
    a = sim_user_40_m[sim_user_40_m.index==user].values
    b = a.squeeze().tolist()
    c = final_policy.loc[:,item]
    d = c[c.index.isin(b)]
    f = d[d.notnull()]
    avg_user = Mean.loc[Mean['user_id'] == user,'rating'].values[0]
    index = f.index.values.squeeze().tolist()
    corr = similarity_with_policy.loc[user,index]
    fin = pd.concat([f, corr], axis=1)
    fin.columns = ['avg_score','correlation']
    fin['score']=fin.apply(lambda x:x['avg_score'] * x['correlation'],axis=1)
    nume = fin['score'].sum()
    deno = fin['correlation'].sum()
    final_score = avg_user + (nume/deno)
    return final_score

#this function is used for finding out the score that user will buy the item or not using the top 30 nearest neighbour of him
```

```python
def User_item_score1(user):
    #fingout the policies rating which the user given
    Policy_seen_by_user = check_table.columns[check_table[check_table.index==user].notna().any()].tolist()
    #finding out the top 40 neighbour of the user
    a = sim_user_40_m[sim_user_40_m.index==user].values
    #squeezing the vlaues of a into scalar form and converting into list
    b = a.squeeze().tolist()
    #getting the policies 40 top neighbour user have 
    d = Policy_user[Policy_user.index.isin(b)]
    #joing the values of the d
    l = ','.join(d.values)
    Policy_seen_by_similar_users = l.split(',')
    #getting the policies which user id we have requested dont have 
    Policys_under_consideration = list(set(Policy_seen_by_similar_users)-set(list(map(str, Policy_seen_by_user))))
    Policys_under_consideration = list(map(int, Policys_under_consideration))
    
    #now finding out the score of the policies whihc the user don't have and using the similarity b/w the policies, and the correlation 
    # b/w the policies and saving it inot the dataframe 
    score = []
    for item in Policys_under_consideration:
        c = final_policy.loc[:,item]
        d = c[c.index.isin(b)]
        f = d[d.notnull()]
        avg_user = Mean.loc[Mean['user_id'] == user,'rating'].values[0]
        index = f.index.values.squeeze().tolist()
        corr = similarity_with_policy.loc[user,index]
        fin = pd.concat([f, corr], axis=1)
        fin.columns = ['avg_score','correlation']
        fin['score']=fin.apply(lambda x:x['avg_score'] * x['correlation'],axis=1)
        nume = fin['score'].sum()
        deno = fin['correlation'].sum()
        final_score = avg_user + (nume/deno)
        score.append(final_score)
    
    data = pd.DataFrame({'policy_id':Policys_under_consideration,'score':score})
    # recommending upto top 5 policies user can buy 
    top_5_recommendation = data.sort_values(by='score',ascending=False).head(5)
    Policy_Name = top_5_recommendation.merge(policy_data, how='inner', on='policy_id')
    Policy_Names = Policy_Name.policy_title.values.tolist()
    return Policy_Names
```
